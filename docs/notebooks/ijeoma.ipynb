```python
# Install the necessary libraries
!pip install geemap

import geemap
import ee

# Authenticate and Initialize the Earth Engine module
ee.Authenticate()
ee.Initialize()

# Define the region of interest (Tshwane, South Africa)
roi = ee.Geometry.Polygon(
    [[[28.0000, -25.0000],
      [28.0000, -26.0000],
      [29.0000, -26.0000],
      [29.0000, -25.0000]]])

# Fetch the MODIS LST data for 2022-23
modis_lst = ee.ImageCollection('MODIS/006/MOD11A2') \
    .filterDate('2022-01-01', '2023-01-01') \
    .select('LST_Day_1km') \
    .mean() \
    .clip(roi)

# Convert the LST from Kelvin to Celsius
lst_celsius = modis_lst.multiply(0.02).subtract(273.15)

# Define visualization parameters
vis_params = {
    'min': 20,
    'max': 40,
    'palette': ['blue', 'green', 'yellow', 'red']
}

# Create a map
Map = geemap.Map(center=[-25.75, 28.5], zoom=10)
Map.addLayer(lst_celsius, vis_params, 'LST 2022-23')
Map.addLayerControl()
Map
```

```python


import geemap
import ee

# Authenticate and Initialize the Earth Engine module
ee.Authenticate()
ee.Initialize()

# Define the coordinates for the Stanza Bopape Clinic (approximate location)
clinic_coords = [28.2940, -25.7174]

# Define a small region of interest (e.g., 1km buffer around the clinic)
roi = ee.Geometry.Point(clinic_coords).buffer(20000)  # Buffer in meters

# Fetch the Landsat 8 Surface Reflectance data for 2022-23
landsat = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
    .filterBounds(roi) \
    .filterDate('2022-01-01', '2023-01-01') \
    .filter(ee.Filter.lt('CLOUD_COVER', 10)) \
    .median() \
    .clip(roi)

# Calculate NDVI using the NIR (Band 5) and Red (Band 4) bands
ndvi = landsat.normalizedDifference(['SR_B5', 'SR_B4']).rename('NDVI')

# Calculate LST using the Thermal Band (Band 10)
# Convert the thermal band to Top of Atmosphere Brightness Temperature (in Kelvin)
lst = landsat.select('ST_B10').multiply(0.00341802).add(149.0)

# Convert LST from Kelvin to Celsius
lst_celsius = lst.subtract(273.15).rename('LST_Celsius')

# Define visualization parameters for NDVI and LST
ndvi_vis_params = {
    'min': -1,
    'max': 1,
    'palette': ['blue', 'white', 'green']
}

lst_vis_params = {
    'min': 20,
    'max': 45,
    'palette': ['blue', 'green', 'yellow', 'red']
}

# Create a map centered around the clinic
Map = geemap.Map(center=[-25.7174, 28.2940], zoom=14)

# Add NDVI and LST layers to the map
Map.addLayer(ndvi, ndvi_vis_params, 'NDVI 2022-23')
Map.addLayer(lst_celsius, lst_vis_params, 'LST 2022-23')
Map.addLayerControl()

# Display the map
Map
```

```python


import geemap
import ee

# Authenticate and Initialize the Earth Engine module
ee.Authenticate()
ee.Initialize()

# Define the coordinates for the Stanza Bopape Clinic (approximate location)
clinic_coords = [28.2940, -25.7174]

# Define a small region of interest (e.g., 1km buffer around the clinic)
roi = ee.Geometry.Point(clinic_coords).buffer(30000)  # Buffer in meters

# Fetch the Landsat 8 Surface Reflectance data for 2022-23
landsat = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
    .filterBounds(roi) \
    .filterDate('2022-01-01', '2023-01-01') \
    .filter(ee.Filter.lt('CLOUD_COVER', 10)) \
    .median() \
    .clip(roi)

# Calculate NDVI using the NIR (Band 5) and Red (Band 4) bands
ndvi = landsat.normalizedDifference(['SR_B5', 'SR_B4']).rename('NDVI')

# Calculate NDBI using the SWIR (Band 6) and NIR (Band 5) bands
ndbi = landsat.normalizedDifference(['SR_B6', 'SR_B5']).rename('NDBI')

# Calculate LST using the Thermal Band (Band 10)
# Convert the thermal band to Top of Atmosphere Brightness Temperature (in Kelvin)
lst = landsat.select('ST_B10').multiply(0.00341802).add(149.0)

# Convert LST from Kelvin to Celsius
lst_celsius = lst.subtract(273.15).rename('LST_Celsius')

# Define visualization parameters for NDVI, NDBI, and LST
ndvi_vis_params = {
    'min': -1,
    'max': 1,
    'palette': ['blue', 'white', 'green']
}

ndbi_vis_params = {
    'min': -1,
    'max': 1,
    'palette': ['white', 'yellow', 'brown']
}

# Updated visualization parameters for LST with a redder filter to indicate hotter areas
lst_vis_params = {
    'min': 20,
    'max': 45,
    'palette': ['yellow', 'orange', 'red', 'darkred']
}

# Create a map centered around the clinic
Map = geemap.Map(center=[-25.7174, 28.2940], zoom=14)

# Add NDVI, NDBI, and LST layers to the map
Map.addLayer(ndvi, ndvi_vis_params, 'NDVI 2022-23')
Map.addLayer(ndbi, ndbi_vis_params, 'NDBI 2022-23')
Map.addLayer(lst_celsius, lst_vis_params, 'LST 2022-23 (Hotter Look)')
Map.addLayerControl()

# Display the map
Map
```
